API integration in Python typically involves using the requests library to make HTTP requests to an API. Below, I'll provide a step-by-step guide on how to integrate with a RESTful API using Python, including examples for GET, POST, PUT, and DELETE requests.
1. Install the requests Library

If you haven't already installed the requests library, you can do so using pip:

```python
# pip install requests
```

2. Import the requests Library

Start by importing the requests library in your Python script

```python
import requests
```

3. Making API Requests
GET Request

A GET request is used to retrieve data from an API.

```python
response = requests.get('https://jsonplaceholder.typicode.com/posts')
if response.status_code == 200:
    data = response.json()
    # print(data)
    for post in data:
    # for post in data[:3]:
        # print(f"Title: {post['title']}")
        # print(post)
        print('Id is  : ', post['id'],'\n','Body :',post['body'] )
        print('\n')
else:
    print(f"Failed to retrieve data: {response.status_code}")
```

4. POST Request

A POST request is used to send data to an API, often to create a new resource.

```python
url = 'https://jsonplaceholder.typicode.com/posts'
payload = {
    'title': 'foo',
    'body': 'bar',
    'userId': 1
}
response = requests.post(url, json=payload)
if response.status_code == 201:
    print('Post created successfully.')
    print(response.json())
else:
    print(f"Failed to create post: {response.status_code}")
```

5. PUT Request

A PUT request is used to update an existing resource.

```python
url = 'https://jsonplaceholder.typicode.com/posts/1'
payload = {
    'id': 1,
    'title': 'updated title',
    'body': 'updated body',
    'userId': 1
}
response = requests.put(url, json=payload)
if response.status_code == 200:
    print('Post updated successfully.')
    print(response.json())
else:
    print(f"Failed to update post: {response.status_code}")
```

6. DELETE Request

A DELETE request is used to delete an existing resource.

```python
url = 'https://jsonplaceholder.typicode.com/posts/1'
response = requests.delete(url)
if response.status_code == 200:
    print('Post deleted successfully.')
else:
    print(f"Failed to delete post: {response.status_code}")
```
